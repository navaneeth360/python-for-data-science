```python
rep2=rep[['State','District','Population','Literate population ','Male population','Male literate population',
          'Female population','Female literate population ']]
rep2
```

```python
rep_dist=rep2[['District','Population','Literate population ','Male population','Male literate population',
          'Female population','Female literate population ']].groupby(['District']).sum()
rep_dist
```

```python
rep_dist['Literacy Rate']=rep_dist['Literate population ']/rep_dist['Population']
rep_dist['Male Literacy Rate']=rep_dist['Male literate population']/rep_dist['Male population']
rep_dist['Female Literacy Rate']=rep_dist['Female literate population ']/rep_dist['Female population']
```

```python
rep_dist
```

```python
rep_dist[rep_dist['Literacy Rate']==rep_dist['Literacy Rate'].max()]
```
